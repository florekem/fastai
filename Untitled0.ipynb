```
def get_data():
    path = datasets.download_data(MNIST_URL, ext='.gz')
  
    with gzip.open(path, 'rb') as f:
        ((x_train, y_train), (x_valid, y_valid), _) = pickle.load(f, encoding='latin-1')
    return map(tensor, (x_train,y_train,x_valid,y_valid))

```

```
x_train,y_train,x_valid,y_valid = get_data()

x_train_mean = x_train.mean()
x_train_std = x_train.std()

x_train_mean, x_train_std
```

```
def normalize(matrix, mean, std): 
  return (matrix - mean) / std
```

```
x_train_normalized = normalize(x_train, x_train_mean, x_train_std)

x_train_normalized.mean(), x_train_normalized.std()
```

```
# NB: Use training, not validation mean for validation set
x_valid_normalized = normalize(x_valid, x_train_mean, x_train_std)
x_valid_normalized.mean(), x_valid_normalized.std()
```

```
x_train_rows_no, x_train_cols_no = x_train.shape

x_train_rows_no, x_train_cols_no
```

```
no_of_hidden_layers = 50
```

```
w1 = torch.randn(x_train_cols_no, no_of_hidden_layers) * math.sqrt(2 / x_train_cols_no) # kaiming init / he init for relu
w2 = torch.randn(no_of_hidden_layers, 1) / math.sqrt(no_of_hidden_layers)

b1 = torch.zeros(no_of_hidden_layers)
b2 = torch.zeros(1)
```

```
def lin(matrix, weights, bias):
  return matrix@weights + bias
```

```
def relu(matrix):
  return matrix.clamp_min(0.) - 0.5
```

```
def model():
  l1 = lin(x_valid_normalized, w1, b1)
  l2 = relu(l1)
  l3 = lin(l2, w2, b2)
  return l3
```

```
def mse(output_of_model, target):
  return (output_of_model.squeeze(-1) - target).pow(2).mean()
```

```
class Model(nn.Module):
    def __init__(self, n_in, nh, n_out):
        super().__init__()
        self.layers = [nn.Linear(n_in,nh), nn.ReLU(), nn.Linear(nh,n_out)]
        self.loss = mse
        
    def __call__(self, x, targ):
        for l in self.layers: x = l(x)
        return self.loss(x.squeeze(), targ)

# klasa zwraca loss czyli cO
```

```
class Model(nn.Module):
    def __init__(self, n_in, nh, n_out):
        super().__init__()
        self.layers = [nn.Linear(n_in,nh), nn.ReLU(), nn.Linear(nh,n_out)]
        
    def __call__(self, x):
        for l in self.layers: x = l(x)
        return x
```

```
def log_softmax(x):
  return (x.exp()/(x.exp().sum(-1,keepdim=True))).log()
```
